Q1.What is the benefit of regular expressions?
A.Regular expressions are a powerful and flexible tool for working with text data in a wide range of applications. Here are some of the main benefits of using regular expressions:
1.Pattern matching: Regular expressions provide a way to match patterns in text data, including complex patterns involving         multiple characters, wildcards, and special characters. This makes it easy to search for and extract specific pieces of         information from large volumes of text.
2.Flexibility: Regular expressions can be used in a wide variety of programming languages and environments, and can be adapted     to many different types of text data. They can also be combined with other tools and techniques for text processing and         analysis, such as tokenization, stemming, and sentiment analysis.
3.Efficiency: Regular expressions can be very fast and efficient, especially when working with large volumes of text data. They   can often perform complex pattern matching and string manipulation tasks much more quickly than manual methods or other text     pr ocessing tools.
4.Standardization: Regular expressions provide a standardized syntax and set of rules for pattern matching, which makes it         easier for programmers and data analysts to share and reuse code across different projects and applications.
5.Validation: Regular expressions can be used to validate input data, such as email addresses, phone numbers, and other types of   user input. This can help improve the quality and reliability of applications that rely on user input data.

Q7.What is the difference between using a vertical bar (|) as an alteration and using square brackets as a character set?
A.Both vertical bar (|) and square brackets ([]) are used to specify alternatives in regular expressions, but they have different meanings and uses.
  The vertical bar (|) is used to separate two or more alternate patterns that can match the same string. When the regular expression engine encounters a vertical bar, it tries to match the left-side pattern first. If it fails to find a match, it tries to match the right-side pattern. For example, the regular expression cat|dog matches either "cat" or "dog".
  On the other hand, square brackets ([]) are used to specify a set of characters that can match a single character in the input string. The characters inside the brackets are treated as a single unit, and the regular expression engine tries to match any one of the characters in the set. For example, the regular expression [abc] matches either "a", "b", or "c".
  One key difference between the two is that vertical bar (|) separates entire patterns, while square brackets ([]) specify individual characters. Vertical bar allows for more complex patterns, while square brackets are used for simple character sets. Additionally, square brackets can include character ranges, such as [a-z], which matches any lowercase letter, or [0-9], which matches any digit.To summarize, vertical bar (|) is used to specify alternate patterns, while square brackets ([]) are used to specify character sets.
