Q1.Define the relationship between a class and its instances. Is it a one-to-one or a one-to-many partnership, for example?
A.In object-oriented programming, a class is a blueprint or a template for creating objects, and instances are the individual objects created from that class. The relationship between a class and its instances is a one-to-many relationship, where one class can have many instances.
  The class defines the attributes and methods that the instances will have, and each instance has its own set of values for those attributes. When you create an instance of a class, you are creating a unique object that has its own identity, separate from other instances of the same class.
  The class serves as a template or a blueprint for creating multiple instances, each with its own unique state and behavior. Changes to the class will affect all instances of that class, but changes to one instance will not affect other instances of the same class.
  
  
Q2.What kind of data is held only in an instance?
A.In object-oriented programming, an instance is an individual object created from a class that contains its own unique data or state. This data is held only in the instance and is separate from data held in other instances of the same class.
  The data held in an instance can include instance variables or attributes, which are unique to that instance and can have different values for each instance. For example, if you have a class representing a person, each instance of that class would have its own name, age, and other personal information stored as instance variables.
  In addition to instance variables, an instance can also hold other data that is specific to that instance and not shared with other instances. This can include any data that is generated or modified during the lifetime of the instance, such as temporary variables used in a method or function, or any data that is dynamically created based on user input or other external factors.
  
  
Q3.What kind of knowledge is stored in a class?
A.In object-oriented programming, a class is a blueprint for creating objects that define a set of attributes and methods. The attributes of a class define the state of the objects created from it, while the methods define the behavior of those objects. Therefore, a class holds the knowledge of how to create objects with a specific set of attributes and how to perform specific actions or behaviors using those objects. The class defines the structure and behavior of the objects that can be created from it, but the specific values of the attributes are stored only in the instances of the class.


Q4.What exactly is a method, and how is it different from a regular function?
A.In object-oriented programming, a method is a function that is defined inside a class and can be called on an instance of that class. A method is different from a regular function in that it has access to the internal state of the object instance and can modify that state if necessary. When a method is called on an object instance, the object instance is implicitly passed as the first argument (often called self) to the method, allowing the method to operate on the object's internal data.

This means that the variable can only be accessed within the class, and attempts to access it from outside the class will result in an AttributeError. Q7.How do you distinguish between a class variable and an instance variable?
A.In Python, a class variable is a variable that is defined in the class definition and is shared by all instances of the class. It is accessed using the class name, and any changes made to the variable affect all instances of the class.
   On the other hand, an instance variable is a variable that is defined within a method of a class or in the __init__() method of the class, and is specific to each instance of the class. It is accessed using the instance name, and changes made to the variable affect only that instance.
