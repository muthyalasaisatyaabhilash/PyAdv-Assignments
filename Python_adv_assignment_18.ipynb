Q1.Describe the differences between text and binary files in a single paragraph.
A.Text and binary files are two different types of files used to store data on a computer. Text files store data in a human-readable format, using characters such as letters, digits, and symbols to represent text. These files are typically used for storing documents, source code, configuration files, and other types of data that can be read and edited using a text editor or word processor. Text files are also usually small in size and can be easily shared and transferred over the internet.
  In contrast, binary files store data in a non-human-readable format, using binary code consisting of 1s and 0s to represent the data. These files are typically used for storing executable programs, images, videos, audio files, and other types of data that require specific software or hardware to be accessed and interpreted. Binary files are also usually larger in size and may not be easily shared or transferred over the internet, especially in cases where the file format is proprietary or not widely supported.

Q3.What are some of the issues with using binary operations to read and write a Python integer directly to disc?
A.Using binary operations to read and write a Python integer directly to disc can have several issues:
1.Endianness: Different computer architectures use different byte orders for integers, which can lead to data corruption or       incorrect values if the byte order is not accounted for. For example, a little-endian machine stores the least significant       byte first, while a big-endian machine stores the most significant byte first.
2.Portability: Binary files created on one platform may not be compatible with other platforms due to differences in byte order,   word size, or file format. This can lead to compatibility issues and errors when reading or writing binary files across         different systems.
3.Maintenance: Binary files are not human-readable, which can make it difficult to inspect or modify their contents. This can     make it harder to debug issues or modify the file format as requirements change.
4.Versioning: Binary files are often tightly coupled to specific versions of software or hardware, which can make it difficult     to maintain backwards or forwards compatibility as the system evolves. This can lead to versioning issues and                   incompatibilities when trying to read or write binary files across different versions of the system or application.


Q4.Describe a benefit of using the with keyword instead of explicitly opening a file.
A.The with keyword is a context manager in Python that can be used to simplify the process of opening and closing files. When a file is opened using the with keyword, Python automatically handles the closing of the file when the block of code in the with statement completes, regardless of whether an error occurs. This can be beneficial in several ways:
1.Avoiding resource leaks: If a file is opened but not explicitly closed, it can result in a resource leak, as the file remains   open and occupies system resources until the program terminates. Using the with statement ensures that the file is closed       automatically, even if an exception is raised.
2.Simplifying code: The with statement can simplify code by reducing the amount of boilerplate code required to handle file I/O.   This can improve readability and make code easier to maintain.
3.Improving code safety: Because the with statement automatically closes the file, it can help to prevent errors caused by         forgetting to close the file explicitly, which can result in data corruption or other issues.
4.Enabling better error handling: When a file is opened using the with statement, any exceptions that occur during file I/O can   be caught and handled within the with block, making it easier to debug and handle errors.

Q8.When is pickling the best option?
A.Pickling is a useful technique for serializing Python objects into a stream of bytes that can be saved to disk or transmitted over a network. This allows the objects to be easily reconstructed later, preserving their state and behavior. Pickling is a good option in a number of scenarios:
1.Storing and loading application state: If your application has a lot of state that needs to be persisted between runs,           pickling is a convenient way to save that state to disk and reload it later.
2.Sending data over a network: Pickling is a convenient way to serialize objects for transmission over a network. It is often     used in distributed computing applications where objects need to be passed between nodes.
3.Caching: Pickling can be used to cache the results of expensive computations. By pickling the results and storing them on       disk, subsequent runs of the application can avoid the expensive computation and load the cached results instead.
4.Configuration management: Pickling can be used to store and load configuration data for an application. This can be useful for   complex applications with many configuration options that need to be saved and loaded.
