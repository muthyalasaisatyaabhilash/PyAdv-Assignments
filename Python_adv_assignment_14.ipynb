Q9.How do you distinguish between a wrapper, a wrapped feature, and a decorator?
A.In programming, the terms "wrapper", "wrapped feature", and "decorator" are related concepts that can sometimes be used interchangeably, but they do have distinct meanings.
  A wrapper is a piece of code that provides a simplified interface to a more complex feature or system. It "wraps" the complexity behind an easy-to-use interface. A wrapper can also provide additional functionality beyond the original feature, such as logging or error handling. The term "wrapper" can be used in different contexts, but typically refers to a software component that provides a layer of abstraction over another component or system.
  A wrapped feature is the original feature or system that is being wrapped by the wrapper. It is the underlying functionality that the wrapper is simplifying or enhancing. For example, a logging wrapper might provide an interface for logging messages to a file, but the actual logging functionality is provided by a separate logging library that the wrapper is wrapping.
  A decorator is a special type of wrapper in Python that is used to modify the behavior of functions or classes. A decorator is a function that takes another function or class as input, modifies it in some way, and then returns the modified version. Decorators are commonly used for tasks like adding logging or timing functionality to functions, or adding authentication or caching to API endpoints.
  
  
Q10.If a function is a generator function, what does it return?
A.A generator function in Python returns a generator object when called. A generator object is a type of iterable, similar to a list or tuple, but instead of storing all of its values in memory at once, it generates values on the fly as they are needed.
  When a generator function is called, it does not execute the function body immediately. Instead, it returns a generator object that can be used to iterate over the values produced by the generator. The generator function only begins executing when the next value is requested from the generator object, either using the next() function or in a loop.
  Generator functions in Python are defined using the yield keyword instead of return. When a generator function encounters a yield statement, it "yields" a value to the caller and temporarily suspends execution of the function, preserving its state. The next time the function is called, it resumes execution from where it left off, picking up where it left off and continuing to generate values until there are no more values to yield.

  A yield statement in a function turns it into a generator function, which produces a sequence of values on the fly as they are needed, instead of computing and returning all of them at once. When a yield statement is encountered, the current state of the function is saved and the yielded value is returned to the caller. The next time the function is called, it resumes execution from where it left off, picking up where it left off and continuing to generate values until there are no more values to yield.
  
  
Q12.Identify at least one benefit of generators.
A.One of the main benefits of generators in Python is their memory efficiency. Because generators produce values on the fly as they are needed, they do not need to store all of the generated values in memory at once, unlike lists or other sequences. This means that generators are particularly useful when dealing with large datasets or infinite sequences that cannot be stored in memory all at once.
  Another benefit of generators is their ease of use and readability. They allow for a more concise and elegant code syntax, especially when working with complex data transformations or iterative algorithms
